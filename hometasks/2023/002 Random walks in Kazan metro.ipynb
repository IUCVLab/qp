# Intro

Random walks are processes, when at each step an agent selects movement direction randomly from the given set of choices. [The simplest random walk process is moving left and right in an integer axis](https://en.wikipedia.org/wiki/Random_walk).
![1-d random walk](https://upload.wikimedia.org/wikipedia/commons/7/76/Random_Walk_example.png)

It has a counter-intuitive property, that an expected drift after $N$ steps is $\sqrt{N}$.

# Context

You will model random walks on a limited axis: Kazan metro. Your agent will hop on a train in a random direction, ride only one station, and then hop off. And then repeat.

![image.png](https://upload.wikimedia.org/wikipedia/commons/thumb/6/6b/%D0%9A%D0%B0%D1%80%D1%82%D0%B0_%D0%BC%D0%B5%D1%82%D1%80%D0%BE_%D0%BA%D0%B0%D0%B7%D0%B0%D0%BD%D0%B8_%D1%83%D0%BB%D1%83%D1%87%D1%88.png/360px-%D0%9A%D0%B0%D1%80%D1%82%D0%B0_%D0%BC%D0%B5%D1%82%D1%80%D0%BE_%D0%BA%D0%B0%D0%B7%D0%B0%D0%BD%D0%B8_%D1%83%D0%BB%D1%83%D1%87%D1%88.png).

Currently we have 11 stations with the central station "**Ploshchad Tukaya**".

```python
import networkx as nx
kazan_metro = [
    "Aviastroitelnaya",
    "Severny Vokzal",
    "Yashlek",
    "Kozya Sloboda",
    "Kremlyovskaya",
    "Ploshchad Tukaya",
    "Sukonnaya Sloboda",
    "Ametyevo",
    "Gorki",
    "Prospekt Pobedy",
    "Dubravnaya",
]
kazan_metro_adjacency = [(kazan_metro[i], kazan_metro[i+1]) for i in range(len(kazan_metro) - 1)]
g = nx.Graph()
g.add_edges_from(kazan_metro_adjacency)
nx.draw_networkx(g)
```

```python
nx.adjacency_matrix(g).todense()
```
